Cmo copiar arreglos numpy?

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4])
```

```python
arr1.shape
```

```python
arr2 = arr1 # es una copia que apunta al mismo espacio de memoria
```

```python
print(arr1)
print(arr2)
```

```python
arr1[0] = 100
```

```python
arr1
```

```python
print(arr1)
print(arr2)
```

```python
arr1 = np.array([1, 2, 3, 4])
# Este es el metodo correcto para copiar arreglos
arr2 = arr1.copy()
arr1[0] = 100
print(arr1)
print(arr2)
```

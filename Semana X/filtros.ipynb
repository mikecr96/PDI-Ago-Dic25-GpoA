## Filtro Laplaciano

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
lena_gray = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic25/Grupo A/images/lena.jpg", 0)
```

```python
np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
], dtype=np.int32)
```

```python
kernel1 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
], dtype=np.int32)
```

```python
kernel1.sum()
```

```python
kernel2 = kernel1 * -1
```

```python
kernel3 = np.array([
    [1, 1, 1],
    [1, -8, 1],
    [1, 1, 1]
], dtype=np.int32)
```

```python
kernel4 = kernel3 * -1
```

```python
lena_gray_k1 = cv.filter2D(lena_gray, -1, kernel1)
```

```python
def show_img(img):
    plt.imshow(img, cmap="gray")
    plt.show()  
```

```python
show_img(lena_gray_k1)
```

```python
lena_gray_k2 = cv.filter2D(lena_gray, -1, kernel2)
show_img(lena_gray_k2)
```

```python
lena_gray_k3 = cv.filter2D(lena_gray, -1, kernel3)
show_img(lena_gray_k3)
```

```python
lena_gray_k4 = cv.filter2D(lena_gray, -1, kernel4)
show_img(lena_gray_k4)
```

## Gaussian Blur

```python
coef_x = cv.getGaussianKernel(11, 0) # No lo vamos a utilizar (de momento) - SON COEFICIENTES
coef_y = cv.getGaussianKernel(11, 0) # No lo vamos a utilizar (de momento) - SON COEFICIENTES
```

```python
coef_x.shape, coef_y.shape
```

```python
coef_x.shape, coef_y.T.shape
```

```python
kernel_gaussiano = coef_x @ coef_y.T
```

```python
show_img(cv.filter2D(lena_gray, -1, kernel_gaussiano))
```

```python
show_img(cv.GaussianBlur(lena_gray, (7, 7), sigmaX=3, sigmaY=2))
```

```python
colores = ["b", "g", "r"]

for i in range(3):
    histograma = cv.calcHist([lena_bgr], [i], None, [256], [0, 255])
    plt.plot(histograma, colores[i])
    plt.title(f"Histograma del canal {colores[i]}")
    plt.show()
```

```python
colores = ["b", "g", "r"]

for n, i in enumerate(colores):
    histograma = cv.calcHist([lena_bgr], [n], None, [256], [0, 255])
    plt.plot(histograma, i)
    plt.title(f"Histograma del canal {i}")
    plt.show()
```

```python
for n, i in enumerate(colores):
    histograma = cv.calcHist([lena_bgr], [n], None, [256], [0, 255])
    plt.plot(histograma, i)
    plt.title(f"Histograma de los canales BGR")
plt.show()
```

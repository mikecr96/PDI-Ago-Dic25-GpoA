```python
filas, columnas = lena_gray.shape
potencia = 2
lena_power = np.zeros(lena_gray.shape, dtype=np.uint8)

for f in range(filas):
    for c in range(columnas):
        lena_power[f, c] = lena_gray[f, c] ** potencia
```

```python
lena_power.dtype
```

```python
plt.imshow(lena_power, cmap='gray')
plt.show()
```

```python
# s = L - 1 - r
# s = 255 - r

# img_copia = img_b.copy()
img_copia = np.zeros(img_b.shape, dtype=np.uint8)

for f in range(num_filas):
    for c in range(num_columnas):
        img_copia[f, c] = 255 - img_b[f, c]
```

```python
plt.imshow(img_copia, cmap='gray')
plt.show()
```

```python
plt.imshow(255 - img_b, cmap='gray')
plt.show()
```

```python
'''
s = 255 - r

cuando r = 255
s = 255 - 255
s = 0

cuando r = 0
s = 255 - 0
s = 255

cuando r = 127
s = 255 - 127
s = 128

cuando r = 200
s = 255 - 200
s = 55
'''
```

```python
# bitwise_not
plt.imshow(cv.bitwise_not(img_b), cmap='gray')
plt.show()
```

```python
# bitwise_not
plt.imshow(np.bitwise_not(img_b), cmap='gray')
plt.show()
```

## Transformación de umbralización

```python
'''
k = 127
si r es < k : s = 0
si r es >= k : s = 255
'''

img_copia2 = np.zeros(img_b.shape, dtype=np.uint8)
k = 127

for f in range(num_filas):
    for c in range(num_columnas):
        if img_b[f, c] >= k:
            img_copia2[f, c] = 255
        else:
            img_copia2[f, c] = 0
```

```python
plt.imshow(img_copia2, cmap='gray')
plt.title('Imagen en blanco y negro')
plt.show()
```

```python
# s = c log10(1 + r)
c = 1
img_log = np.zeros(img_b.shape, dtype=np.uint8)

for x in range(num_filas):
    for y in range(num_columnas):
        img_log[x, y] = c * np.log10(1 + img_b[x, y].item())
```

```python
plt.imshow(img_log, cmap="gray")
plt.show()
```

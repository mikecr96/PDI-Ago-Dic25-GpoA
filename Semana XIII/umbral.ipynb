```python
def showImg(img, title=""):
    plt.imshow(img, cmap="gray", vmin=0)
    plt.title(title)
    plt.show()
```

### Búsqueda de un mejor umbral

OTSU

```python
# Con OpenCV 
# umbral, ardilla_bin3 = cv.threshold(ardilla_gris, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
umbral, ardilla_bin3 = cv.threshold(ardilla_gris, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)

showImg(ardilla_bin3, umbral)
```

TRIANGLE

```python
# Con OpenCV 
umbral, ardilla_bin3 = cv.threshold(ardilla_gris, 0, 255, cv.THRESH_BINARY | cv.THRESH_TRIANGLE)

showImg(ardilla_bin3, umbral)
```

```python
sudoku_g = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic25/Grupo A/images/sudoku.png", 0)
showImg(sudoku_g)
```

## FOR

```python
# for (i=0; 100; i++)
for i in lista:
    print(i)
print("hola mundo")
```

```python
for letra in "Hola mundo":
    print(letra)
```

```python
# for (i=0; 100; i++)
for i in range(1, 11, 2):
    print(i)
```

```python
for contador in range(100):
    print("Hola mundo")
```

## While

```python
contador = 0

while contador != 100:
    # contador = contador + 1
    contador += 1
    print(contador, "Hola mundo")
```

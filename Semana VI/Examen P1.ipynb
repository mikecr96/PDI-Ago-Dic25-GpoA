## Parte 3: Transformaciones de Intensidad (25 puntos)

### Ejercicio 3.1 - Transformación Negativo (8 puntos)
Implemente la transformación negativo s = L - 1 - r, donde L=256 para imágenes de 8 bits.
Aplique esta transformación a la imagen en escala de grises y muestre el resultado.

```python
def transformacion_negativo(imagen):
    """
    Aplica transformación negativo a una imagen
    """
    # Su código aquí
    return 256-1-imagen

# Aplicar transformación y mostrar resultado
imagen_gray = cv.imread("/Users/miguelcamargorojas/Downloads/meme.jpg", 0)
imagen_gray_negativo = transformacion_negativo(imagen_gray)
plt.imshow(imagen_gray_negativo, cmap="gray")
plt.show()
```

### Ejercicio 3.2 - Transformación Logarítmica (8 puntos)
Implemente la transformación logarítmica s = c * log(1 + r), donde c es una constante.
Use c = 255 / np.log(1 + np.max(imagen)). Aplique esta transformación y muestre el resultado.

```python
def transformacion_logaritmica(imagen):
    """
    Aplica transformación logarítmica a una imagen
    """
    # Su código aquí
    c = 255 / np.log10(1 + np.max(imagen).item())
    s = c * np.log10(1 + imagen)
    
    return s

# Aplicar transformación y mostrar resultado
imagen_gray_log = transformacion_logaritmica(imagen_gray)
plt.imshow(imagen_gray_log, cmap="gray")
plt.show()
```

### Ejercicio 3.3 - Corrección Gamma (9 puntos)
Implemente la transformación gamma s = c * r^γ, donde c es una constante de normalización.
Creé una función que permita ajustar el valor de gamma y muestre los resultados para γ = 0.5, γ = 1.0, y γ = 2.0.

```python
def correccion_gamma(imagen, gamma):
    """
    Aplica corrección gamma a una imagen
    """
    # Su código aquí
    
    return imagen ** gamma

# Aplicar corrección gamma con diferentes valores
gammas = [0.25, 0.5, 1.0, 1.5, 2.0]
imgs_gama = []
for gamma in gammas:
    imgs_gama.append(correccion_gamma(imagen_gray, gamma))

# Su código para mostrar los resultados
for contador, img in enumerate(imgs_gama):
    plt.imshow(img, cmap="gray")
    plt.title(f"Transformación con gamma {gammas[contador]}")
    plt.show()
```

## Parte 4: Histogramas (25 puntos)

### Ejercicio 4.1 - Histograma en Escala de Grises (10 puntos)
Calcule y visualice el histograma de la imagen en escala de grises.
Muestre tanto la imagen como su histograma lado a lado.

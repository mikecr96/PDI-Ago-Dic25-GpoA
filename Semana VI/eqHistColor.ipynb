```python
# Esto NO viene en el examen del Primer parcial

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
lena_bgr = cv.imread('/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic25/Grupo A/images/lena.jpg')
lena_rgb = lena_bgr[:, :, ::-1]
```

```python
lena_bgr.shape
```

```python
plt.imshow(lena_rgb)
plt.show()
```

```python
"""
1) Separar canales
2) Ecualizar el/los canal(es) deseados
3) Unir canales
"""

canal_r = lena_rgb[:, :, 0]
canal_g = lena_rgb[:, :, 1]
canal_b = lena_rgb[:, :, 2]
```

```python
plt.imshow(canal_r, cmap="gray")
plt.show()
```

```python
hist_r = cv.calcHist([canal_r], [0], None, [256], [0, 256])
```

```python
plt.plot(hist_r)
plt.title("Histograma del canal R")
plt.show()
```

```python
# Ecualizamos el canal R
canal_r_eq = cv.equalizeHist(canal_r)
canal_g_eq = cv.equalizeHist(canal_g)
canal_b_eq = cv.equalizeHist(canal_b)
```

```python
hist_r_eq = cv.calcHist([canal_r_eq], [0], None, [256], [0, 256])
plt.plot(hist_r_eq)
plt.title("Histograma del canal R_eq")
plt.show()
```

```python
# Paso 3) Unir
lena_rgb_req = cv.merge([canal_r_eq, canal_g, canal_b])
```

```python
lena_rgb_eq = cv.merge([canal_r_eq, canal_g_eq, canal_b_eq])
```

```python
plt.imshow(lena_rgb_req)
plt.show()
```

```python
plt.imshow(lena_rgb_eq)
plt.show()
```
